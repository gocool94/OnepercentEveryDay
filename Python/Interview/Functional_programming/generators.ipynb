**Generators and Iterators:**

1. Write a generator to yield Fibonacci numbers up to a given limit.
2. Implement a generator to generate the prime numbers indefinitely.
3. Create a generator to iterate through the lines of a text file.
4. Write a custom iterator to iterate through the keys of a dictionary.

```python
def simpleGeneratorFun():
    yield 1           
    yield 2           
    yield 3           
  
# Driver code to check above generator function
for value in simpleGeneratorFun():
    print(value)
```

```python
def fibonocci(limit):
    a = 0
    b = 1

    while(a<=limit):
        yield a
        a,b = b,a+b


x = fibonocci(100)
for num in x:
    print(num)

```

```python
# 2 . prime numbers

def prime_number(limit):

    for i in range(2,limit):
        init = 2
        num = i
        while(init<=(num//2)):
            if (num%init==0):
                break
            init+=1
        else:
            yield i


primenumbergen = prime_number(100)

for values in primenumbergen:
    print(values,end=" ")
```

```python
"3: iteratre through the lines of the text_file"

def lines_generator(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

# Replace 'path/to/your/text_file.txt' with the actual path to your text file
file_path = 'path/to/your/text_file.txt'
lines_gen = lines_generator(file_path)

for line in lines_gen:
    print(line)



```

```python
"4. custom iterator to iterate through the keys of the dictionary"

def dict_clean(json):
    print(json)
    for i in json:
        yield i 


gen = dict_clean({1:2,2:3,3:4,5:6,7:8})

next(gen)
next(gen)

for i in gen:
    print(i)
```

```python
# 5. Generator to flatten a nested list


def flattern(list):
    for i in list:
        for j in i:
            yield j

list_w = [[1,2,3],[2,3,4]]
generator_value = flattern(list=list_w)

new_list = []
for i in generator_value:
    new_list.append(i)

new_list
```

```python
# 6. Generator to fibonocci for a given limit


def fib(limit):
    a , b = 0 , 1
    while(a <= limit):
        yield a
        a , b = b , a + b
    
new_gen = fib(1000)

for i in new_gen:
    print(i,end=" ",sep= ",")
```
