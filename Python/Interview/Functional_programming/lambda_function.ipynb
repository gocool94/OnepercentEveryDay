# Lambda functions

## Syntax :

    > lambda arguments : expression


```python
x = lambda a,b : a + b + 2

x(2,3)

```

```python
y = lambda num1, num2 : num1 * num2
y(3,4)
```

```python
"""1. Write a lambda function to calculate the square of a number.
"""

square = lambda n : n * n
square(2)
```

```python
"""3. Write a lambda function to check if a number is even.
"""

even = lambda n : n% 2 == 0

print(even(21))
```

```python
"""19. Use a lambda function to find the maximum of two numbers.
"""

greatest = lambda x , y : x if x > y else y 
greatest(2,3)
```

```python
"""20. Write a lambda function to extract the last element from a list.
"""
lastelement = lambda l : l[-1]

lastelement([1,2,3,4,5])
```

```python
"""3.  Write a program to find the cube of each element in a list using a lambda function.
"""

cube = list(map(lambda x: x **3,[2,3,4,5]))
cube
```

```python
data = [(1, 5), (2, 3), (3, 8), (4, 1)]
#  Use a lambda function to sort a list of tuples based on the second element.

list = sorted(data,key=lambda x : x[1])
list

```

```python
#2.  Implement a program to sort a list of strings based on their last character using a lambda function.
# Take input from the user
input_strings = input("Enter a list of strings separated by spaces: ").split()

# Sort the list based on the last character using a lambda function
sorted_strings = sorted(input_strings, key=lambda x: x[-1])

# Print the sorted list
print("Sorted list based on last character:")
for string in sorted_strings:
    print(string)
```
