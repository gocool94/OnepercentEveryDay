```python
"""4. Write a list comprehension to flatten a 2D list into a 1D list.
"""

# Original 2D list
two_d_list = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]

# Flattened 1D list using a list comprehension
flattened_list = [element for sublist in two_d_list for element in sublist]

print(flattened_list)
```

```python

def isprime(NumbertoCheck):
    if NumbertoCheck == 0 or NumbertoCheck == 1:
        return False
    num = 2
    while(num<=NumbertoCheck/2):
        if(NumbertoCheck%num == 0):
            return False
            
        num+=1
    else:
        return True

print([num for num in range(0,100+1) if isprime(num)])
```
