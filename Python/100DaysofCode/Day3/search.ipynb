```python
# Binary Search


"""
Given an array arr[] sorted in ascending order of size N and an integer K. Check if K is present in the array or not.
"""

def searchInSorted(arr, k):
    print(arr) # the array
    
    print(k) # Search element

    low = 0
    high = len(arr)-1

    
    while (low <= high):
        
        mid = (low + high)//2
        print(low,mid)
        if (arr[mid] == k):
            return 1
        
        elif(arr[mid] > k  ):
            high = mid -1
        
        elif (arr[mid] < k):
            low = mid + 1
    return -1

print(searchInSorted([1,2,3,4,6],6))



    
        


        

```
