```python
def theSequence(n):
    #code here
    if n == 0 :
        return 1
        
    return n+n*(theSequence(n-1))

print(theSequence(4))
```

```python
def printArrayRecursively(arr,n):
        if(n==0):
            return
        printArrayRecursively(arr,n-1)
        #print(n)
        #print(arr)
        print(arr[n-1], end = " ")

printArrayRecursively(arr = [1,2,3,4,5],n = 5)
```
